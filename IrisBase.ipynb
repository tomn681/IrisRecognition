# Parte 1: Detección de Iris

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

### Load Image

```python
path = '/media/dela/1TB/A Universidad/Electrica/VIII Sem/Procesamiento Digital de Imagenes/IrisRec/IrisRecognition/IrisTestImages/'

image = mpimg.imread(path + 'iris1.jpg')
plt.imshow(image)
```

### Image to Binary

```python
grayImage = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(grayImage, cmap = 'gray')
```

```python
thresh, BinaryImage = cv2.threshold(grayImage, 20, 255, cv2.THRESH_BINARY)
plt.imshow(BinaryImage, cmap = 'gray')
```

### Pupil Detection

```python
def pupilDetector(img, block_size = 3, num_tries = 3):
    blur = cv2.GaussianBlur(img,(5,5), 0) #Gaussian Smoothing, kernel 5x5, standard deviation (x, y) 0
    canny = cv2.Canny(blur, 400, 500) #Algoritmo de Canny para la deteccion de bordes, theresold (Hi/Lo) 100/200
    #canny = cv2.circle(canny, (104, 80), 35, (255, 255, 255), 3)
    return img, blur, canny
img, blur, canny = pupilDetector(BinaryImage)
plt.imshow(canny, cmap = 'gray')
```
